```python
import copy

n = 2500
y = [0]*2
x = [0]*(n+2)
y[0] = 15
m = (pow(2,31)-1)
for i in range(1,n+2):
    y[1] = (16807 * y[0]) % m  # Usar o operador % para calcular o módulo
    x[i] = y[1] / m
        
    y[0] = y[1]


vetor_x = copy.deepcopy(x)
vetor_x.pop()
vetor_y = copy.deepcopy(x)
vetor_y.pop(0)

print(vetor_x)
print(vetor_y)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import csv


def plot_grafico_bidimensional(vetor_x, vetor_y, titulo="Gráfico Bidimensional", label_x="Eixo X", label_y="Eixo Y"):
    """
    Plota um gráfico bidimensional usando dois vetores como eixos.

    :param vetor_x: Vetor de valores para o eixo X.
    :param vetor_y: Vetor de valores para o eixo Y.
    :param titulo: Título do gráfico (opcional).
    :param label_x: Rótulo do eixo X (opcional).
    :param label_y: Rótulo do eixo Y (opcional).
    """
    plt.figure(figsize=(7,7))
    plt.scatter(vetor_x, vetor_y)
    plt.title(titulo)
    plt.xlabel(label_x)
    plt.ylabel(label_y)

    plt.show()
    



plot_grafico_bidimensional(vetor_x, vetor_y, "Meu Gráfico de Exemplo", "Eixo X", "Eixo Y")
```

```python
import numpy as np 
print("Média: ",np.average(vetor_y))
print("Desvio Padrão: ",np.std(vetor_y))
mat = np.array([vetor_x,vetor_y])
print("Covariância: ",np.cov(mat))


```

```python
from scipy.stats import chi2

# Parâmetros
alfa = 0.05
numero_intervalos = 10

# Histograma usando o número de intervalos
hist, bin_edges = np.histogram(vetor_x, bins=numero_intervalos)

# Teste de qui-quadrado
observados = hist
esperados = len(vetor_x) / numero_intervalos

graus_de_liberdade = numero_intervalos - 1

qui_quadrado = np.sum((observados - esperados) ** 2 / esperados)
valor_p = 1 - chi2.cdf(qui_quadrado, graus_de_liberdade)

print("Valor da estatística de qui-quadrado:", qui_quadrado)
print("Valor p:", valor_p)

# Teste de hipóteses
if valor_p < alfa:
    print("Rejeitar a hipótese nula")
else:
    print("Não rejeitar a hipótese nula")
```
