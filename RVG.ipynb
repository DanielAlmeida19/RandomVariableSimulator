<H1>Geração de variáveis aleatórios com distribuição uniforme<H1>

```python
from RNG import Generator
import numpy as np
import matplotlib.pyplot as plt
import math

a=2
b=5
x = Generator(13,2500)

for i in range(0,len(x)): 
    x[i]=(b-a)*x[i]+a

fx = []
for i in x:    
    fx.append(1/(b-a))

print(x)   
print(fx)
plt.scatter(x,fx)
plt.show()
```

<H1>Geração de variáveis aleatórias com distribuição exponencial

```python

alpha = 2
x = Generator(13,2500)

for i in range(1,len(x)): 
    x[i]=-(math.log(x[i])/alpha)
    

def f(x):
    fx = []
    for i in x:
        fx.append(alpha*math.exp(-alpha*i))
    return  fx

fx = f(x)
print(x)
print(fx)
plt.scatter(x,fx)
plt.show()
```

<H1>Geração de variáveis aleatórias com distribuição Weibull

```python
alpha = 1
beta = 0.5
x = Generator(13,2500)
print(x)


for i in range(2,len(x)): 
    x[i]=(-(math.log(x[i])**(1/alpha)))*beta

fx = []
for i in x:
    fx.append(alpha*beta**-alpha*i**(alpha-1)*math.exp(-(i/beta)**alpha))
    
print(x)
print(fx)
plt.scatter(x,fx)  

plt.show()  
```

<H1>Geração de variáveis aleatórias com distribuição Erlang

```python
import copy


k = 3
alpha = 1

y = Generator(13,2500)
x = [0]*len(y)
for i in range(1,len(y)):
    for j in range(1,k+1):
        x[i] += (math.log(y[i])/alpha)
    x[i] = -x[i]

fx = []
for i in x:
    fx.append((alpha*math.exp(-alpha*i)*(alpha*i)**(k-1))/math.factorial(k-1))
    
print(x)
print(fx)
plt.scatter(x,fx)  

plt.show()  
```

<H1>Geração de variáveis aleatórias com distribuição Triangular

```python

a = 1
b = 3
c = 4
x = Generator(13,2500)
for i in range(0,len(x)):
    if(x[i]>= 0 and x[i]<= ((b-a)/(c-a))):
        x[i] = a + (x[i]*(b-a)*(c-a))**(1/2)
    else:
        x[i] = c - ((1-x[i])*(c-b)*(c-a))**(1/2)

fx = []
for i in x:
    if(i >= a and  i<=b):
        fx.append((2*(i-a))/((b-a)*(c-a)))
    elif(i>b and i<=c):
        fx.append((2*(c-i))/((c-b)*(c-a)))
        
    
print(x)
print(fx)
plt.scatter(x,fx)  

plt.show()  
```
